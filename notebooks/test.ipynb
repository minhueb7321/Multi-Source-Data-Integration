```python
# Nhân sự trên DMS
def fetch_dms_employees(file_path):
    df = pd.read_excel(file_path)
    # Lấy các cột cần thiết

    df.columns = [col.capitalize() for col in df.columns]
    df = df[['Mã nhân viên', 'Tên nhân viên', 'Phòng ban', 'Chức danh']]
    return df

nhan_vien = fetch_dms_employees(r'')

nhan_vien.info()
```

```python
import os
os.listdir(r'F:\Documents\etl_project\data\raw\misa\employees')


# for file in os.listdir(r'F:\Documents\etl_project\data\raw\misa\employees'):
def fetch_misa_employees(file_path):
    # Truyền engine xlrd vào 
    df = pd.read_excel(file_path, header=1, engine='xlrd')
    # Loại bỏ dòng cuối
    df = df.iloc[:-1]
    # Lấy các cột cần lấy 
    df  =df[['Mã nhân viên', 'Tên nhân viên']]
    return df


def concatnate_all_employees(folder_name):
    all_files = []
    if len(os.listdir(folder_name)) == 12:
        print("✅ Đủ 12 file nhân sự Misa")
        for file in os.listdir(folder_name):
            if file.endswith('.xls'):
                path_file = os.path.join(folder_name, file)
                df = fetch_misa_employees(path_file)
            all_files.append(df)
        final_df = pd.concat(all_files, ignore_index=True)
        print(f"✅ Xuất file thành công . Tổng số dòng: {len(final_df)}, tổng số cột: {len(final_df.columns)}")
        return final_df
    else:
        print("Không đủ file nhân sự để cập nhật, chú ý update hàng tháng !!")

df = concatnate_all_employees()
    
```
